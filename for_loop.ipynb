## For loops --> iterating everything

### Loop through list

```python
marks = [1,2,3,4,5]

for i in marks:
   print(i)
```

### Loop through string 

```python
name = "anjali"

for ch in name:
    print(ch)
```

## loop through tuple

```python
color = ('red','purple','yellow')

for colors in color:
    print(colors, end=" ")
```

### loops through set 

```python
fruits = {'apple','mango','banana'}

for i in fruits:
    print(i)
 
```

### loop through dict  - 3 ways

```python
students = {'name':'anjali','age':22, 'marks':91}

# way 1 --> key only (default)

for i in students:
    print(i)

# way 2 --> values only 

for i in students.values():
    print(i)

# way 3 --> both key and values 

for i ,a in students.items():
    print(i,a)
```

```python
# default --> get key 
# get value --> students.values()
# get key and value --> students.items()
```

### loop through range 

```python
for i in range(5):
    print(i)

# range with start , and stop 
for i in range(1,11):
    print(i)

# range with step 
for i in range(1,11,2):
    print(i)

# count down
for i in range(10,0,-1):
    print(i)
```

### Enumerate --> index + value together 

```python
students = ['anjali','rahul','om']

for i in range(len(students)):
    print(f"{i+1}.{students[i]}")
```

```python
for i, a in enumerate(students,1):
    print(f"{i}.{a}")
```

```python
students = {"name":"anjali", "age":22}

for i , a in enumerate(students.values()):
    print(f"{i},{a}")  # here i is index number , a is value 
```

```python
### zip --> loop 2 list together 
name = ['anjali','rahul','Priya']
marks =[91,89,90]
grade = ['A','B','A']

for n,m,g in zip(name,marks,grade):
    print(f"{n}-> {m} -> {g}")
```

### nested loop -> loop inside loop

```python
for i in range(1,4):    # outer --> run 3 times
    for j in range(1,4): # inner --> runs 3 times per outer
        print(f"{i} x {j} = {i*j}")
    print("-----")
```

```python
### outer runs once --> inner runs completely -> repeat
```

### break , continue , pass 

```python
## break --> stop loop completely 

for i in range(10):
    if i == 5:
        break 
    print(i)
```

```python
## contine - skip this iteration and continue 

for i in range(10):
    if i == 5:
        continue 
    print(i)
```

```python
### pass - placeholder , do nothing 

for i in range(10):
    if i == 5:
        pass
    print(i)
```

```python
## how loop works 
works = [91,02,90,58]
for i in works:
    iteration 1 -> works = 91 --> run Code
    iteration 2 -> works = 02 --> run code 
    iteration 3 -> works = 90 --> run code 
    iteration 4 -> works = 58 --> run code 
no more items --> loop ends
```

# 10 Problem 

```python
marks = [91, 38, 78, 55, 29, 88, 61]
Loop through with enumerate
Print → "1. 91 → Pass ✅" or "1. 38 → Fail ❌"
```

```python
input --> marks = [91, 38, 78, 55, 29, 88, 61] 

process -> 1_ loop through enumerate 
2 less fail , high pass 

output --> "1. 91 → Pass ✅" or "1. 38 → Fail ❌"
```

```python
marks = [91, 38, 78, 55, 29, 88, 61] 

for i , a in enumerate(marks,1):
    if a > 50:
        print("pass")
    else:
        print("fail")
    print(f"{i}.{a}")
```

```python
marks = [91, 38, 78, 55, 29, 88, 61] 

for i , a in enumerate(marks,1):
    result = "pass" if a >=50 else "Fail"
    print(f"{i}.{a}-->{result}")

```

### problem 3

```python
sentence = "Anjali is learning Python"
Loop through string
Count how many vowels (a,e,i,o,u)
Print → "Total vowels: 8"

i have to count total vowels present in this sentence sentence = "Anjali is learning Python"
input --> sentence = "Anjali is learning Python"

process --> "aeiou" in sentence , count append in total_vowels 
```

```python
total_vowels = 0
sentence = "Anjali is learning Python"

for ch in sentence:
    if ch in ('a','e','i','o','u'):
        total_vowels += 1
print(f"total vowel,{total_vowels}")
```

### 7 Problem 

```python
names  = ["Anjali", "Rahul", "Priya", "Amit"]
marks  = [91, 78, 85, 60]
grades = ["A", "B", "B", "C"]

Using zip + enumerate —
Print → "Rank 1: Anjali | Marks: 91 | Grade: A"


input = names  = ["Anjali", "Rahul", "Priya", "Amit"]
marks  = [91, 78, 85, 60]
grades = ["A", "B", "B", "C"]  # list of data
------------------------------------------------------------

process data = using zip + enumerate 

zip = combine all the list 
enumerate = index + value 



-------------------------------------------------------------
output = Print → "Rank 1: Anjali | Marks: 91 | Grade: A" 
```

```python
names  = ["Anjali", "Rahul", "Priya", "Amit"]
marks  = [91, 78, 85, 60]
grades = ["A", "B", "B", "C"]

for index , (n,m,g) in enumerate(zip(name,marks,grades),1):
    print(f" Rank{index}.{n}|{m},{g}")
```

### 10 - Problem 

```python
words = ["python", "is", "fun", "and", "powerful"]
Loop through list
For each word — loop through its characters
Print word + its character count
Format → "python → 6 characters"
Don't use len() — count manually using loop

## each word loop through each charchter and count each word

```

```python
words = ["python", "is", "fun", "and", "powerful"]


for i in words:   # once per word
    count = 0     # once per word
   
    for j in i:   # once per character
        count += 1
    print(f"{i} -->{count}")   # oncer per word
    
  
```

```python
Outside both loops → one count for the entire list.
Inside outer loop → one count per word. ✅
Inside inner loop → count resets for every character ❌
```
