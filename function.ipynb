```python
### * args --> any numbers 

def total(*args):
    print(type(args))
    print(args)
    return sum(args)

print(total(10,20))
```

### ** kwargs --> dict - key - value pair 

```python
def display(**kwargs):      # key-value pairs
    print(type(kwargs))     # dict
    for key , value in kwargs.items():
        print(f"{key} : {value}")
display(name="anjali",age=23,city='mumbai')
```

```python
### kwargs --> packs all key- value arguments into a dict
### loop through it with --> for key , value in kwargs.items()
```

### *args + **kwargs together 

```python
def show(*args, **kwargs):
    print("args:", args)
    print("kwrags:",kwargs)
show(1,2,3,name="anjali",age=23)
```

### Function inside function 

```python
def outer(marks):
    def get_grade(m):
        if m >= 90: return "A"
        elif m >= 75: return 'B'
        elif m >= 60: return "C"
        elif m >= 40: return "D"
        else: return "F"
    
    for mark in marks:
        grade = get_grade(mark)
        print(f"{mark} --> {grade}")

outer([91,22,80,57,80,34])
```

### P2

```python
Write function called calculate —
takes 2 numbers and operation as parameters
operations: "add", "sub", "mul", "div"
returns result
handle division by zero!
default operation = "add"
```

```python
def calculate(a,b,operation='add'):
   if operation == 'add': return a + b
   elif operation == 'div':
    if b == 0:
       print("cannot be divide by 0")
    else:
       return a // b 
   elif operation == 'mul': return a * b
   else: return a - b 
    
```

```python
print(calculate(2,3))
print(calculate(2,3,'mul'))
print(calculate(2,3,'div'))
print(calculate(2,3,'sub'))
```

### P3 

```python
Write function called analyse_marks —
takes a LIST of marks
returns — highest, lowest, average, pass count, fail count
ALL in one return statement
Print all 5 values neatly
```

```python
def analyse_marks(marks):
    passed = filter(lambda x: x >= 40, marks)
    fail = filter(lambda x : x < 40,marks)
    pass_count = list(passed)
    fail_count = list(fail)
    return max(marks), min(marks), sum(marks)/len(marks), len(pass_count), len(fail_count)
```

```python
analyse_marks([20,10,50,79,50,50,60,70])
```

```python
## gst amount = (gst rate * total amount) / 100
```

```python

def grand_total(*args):
    total =0 
    for arg in args:
        total += arg
    gst = 18 * total / 100 # "Does this calculation depend on the final total, or should it happen for each item?"
    final_amount = total + gst 
    return final_amount
```

```python
## discount = (discount rate / origianl price) / 100
## final = original - discount amount
```

```python
def apply_discount(price,discount):
    discount_amount = (discount * price) / 100
    final_amount = price - discount_amount
    return final_amount
    
```

## P9

Write function called filter_students —
takes list of (name, marks) tuples
takes minimum_marks as parameter (default=40)
returns only students who passed
Use lambda inside function for filtering
Print passed students

```python
def filter_students(students,minimum_marks = 40):
    passed = filter(lambda x:x[1] >= minimum_marks,students)
    passed_std = list(passed)
    return passed_std
```

```python
std = filter_students(students=[  ("Anjali", 91),
    ("Rahul", 38),
    ("Priya", 78),
    ("Amit", 55),
    ("Neha", 29)])
```

```python
Write function called full_report —
takes **kwargs where each key=student name
and value=list of their marks

For each student —
   calculate average (lambda)
   find grade based on average
   print full report

Example call —
full_report(
    Anjali=[91, 85, 78],
    Rahul=[38, 42, 55],
    Priya=[71, 68, 79]
)
```

```python
def full_report(**kwargs):
    average = lambda x : sum(x)/len(x),kwargs.items()
    if average >= 90: return "A"
    elif average >= 80: return "B"
    elif average >= 65: return "C"
    elif average >= 40: return "D"
    else:
        return "F"

print(full_report(name='anjali',marks=[10,30,50,60]))
```

```python
def full_report(**kwargs):
    average = lambda x : sum(x)/len(x)

    for n, m in kwargs.items():
        avg = average(m)

        if avg >= 90: grade = "A"
        elif avg >= 80:
            grade = "B"
        elif avg >= 65:
            grade = "C"
        elif avg >= 40:
            grade = "D"
        else:
            grade = "F"

        print(f"Name    : {n}")
        print(f"Marks   : {m}")
        print(f"Average : {avg:.2f}")
        print(f"Grade   : {grade}")
        print()

full_report(Anjali=[91,30,78],
            Rahul=[89,35,77],
            Priya=[71,68,97])
```
