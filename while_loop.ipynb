## level 1 - basic while loop 

```python
count = 0 

while count < 5:     # condition 
    print(count)     # code to repeat
    count += 1       # move toward stopping 
```

# Three parts always 

```python
1) variable before loop -->  count =0
2) condition --> while count < 5 
3) update inside loop --> count += 1 never forget this without it infinite loop 
```

```python
count = 0

while count < 4:
    # print(count)
    count += 1
print(count)
```

```python
count = 
check 0 < 5 --> True --> print 0 --> count 1
check 1 < 5 --> True --> print 1 --> count 2
check 2 < 5 --> True --> Print 2 --> count 3 
check 3 < 5 --> True --> print 3 --> count 4 
check 4 < 5 --> True --> print 4 --> count 4
check 5 < 5 --> False --> stop 
```

## level 2 --> while True + break 

```python
# when you don't know the condition upfront 

while True:
    age = int(input("enter your age:"))
    if age > 0 and age <= 120:
        print('valid age')
        break 
    else:
        print('invalid age ')
```

```python
while True --> keep goinf forever 
break --> the only exit door 
```

```python
attemts = 0 
max_attemts = 3 

while attemts < max_attemts:
    password = input('enter your password')
    attemts += 1
    if password == "python123":
        print(f"access granted took {attemts} attempts")
        break 
    else:
        remaining = max_attemts - attemts 
        print(f"wrong {remaining} attemts left")
else:
    print("account locked")
```

## level 4 - while with else 

```python
count = 0

while count < 3:
    print(count)
    count += 1
else:
    print("loop done")   # else only run if no break in code
```

```python
else runs  →  when condition becomes False naturally
else SKIPS →  when break is used
```

## level 5 - nested while 

```python
i = 1

while i <= 3:
    j = 1
    while j <= 3:
        print(f"{i} * {j} = {i*j}")
        j+=1
    i+=1
```

## level 6 pattern 1 -- Input validation 

```python
while True:
    marks = int(input('enter marks (0-100)'))
    if 0 <= marks <= 100:
        print('valid')
        break 
    print('INVALID')
print(f'marks {marks}')
```

## Pattern 2 - Retry with limit

```python
attempts = 0 
while attempts < 3:
    pin = input("enter pin:")
    if pin == "12345":
        print('correct')
        break 
    attempts += 1
    print(f'wrong {3-attempts} tires left')
else:
    print('card blocked')
```

## pattern -- manu system 

```python
while True:
    print("1.add 2.View 3.Exit")
    choice = input('choice:')
    if choice == '1':
        print('Adding')
    elif choice == '2':
        print('viewing')
    elif choice == '3':
        print('Goodbye')
        break
    else:
        print('invalid')
```

## pattern 4 -- Accumulator

```python
total = 0
count = 0

while True:
    num = input('enter number(or done):')
    if num =='done':
        break
    total += int(num)
    count += 1
print(f"total: {total}")
print(f"average {total/count:.1f}")
```

```python
For loop                                     while loop

1) for i in range                           1) while count < 5
2) for item in list                         2) while True 
3) i know how many times                    3) i don't know how many 
4) looping through data                     4) waiting for condition 
5) cleaner for collection                   5) better for user input retry logic, game loop
```

## Common mistakes 

```python
# mistake 1 - infinite loop 
count = 0 
while count < 5:
    print(count)
    # forget count += 1 --> runs forever

# mistake 2 - condition never True 
count = 10
while count < 5L # 10<5 = False from start 
    print(count)  # never runs

# mistake 3 - off by one 
count = 0
while count <= 5: ## runs 6 times (0,1,2,3,4,5)
while count < 5: # runs 5 times (0,1,2,3,4)
```

```python
while loop --> keep doing this while condition is true 
               stop when condition becomes false 

while True + break --> loop forever i'll TELL  you when to stop 
        
Parts needed —
1. variable before loop    →  count = 0
2. condition               →  while count < 5
3. update inside           →  count += 1

Real uses —
input validation           →  keep asking until valid
retry logic                →  keep trying until success
menu systems               →  keep showing until exit
games                      →  keep playing until game over

```

###  P10

```python
while True menu —
1. Add marks
2. Show all marks
3. Show average
4. Show pass/fail count
5. Exit

Use list to store marks
Use lambda for calculations
Use filter for pass/fail
Keep running until user chooses 5
```

```python
mark = []
while True:
    # show menu 
    print('\n------------Marks Manager---------')
    print('1. Add marks')
    print('2.Show all marks')
    print('3.Show average')
    print('4.show pass/fail count')
    print('5.Exit')
    choice = input("choice:")

    if choice == "1":
        marks = int(input('enter marks from 0 - 100:'))
        if marks > 0 and marks <= 100:
            mark.append(marks)
        else:
            print(f"{marks} invalid marks")
    
    elif choice == '2':
        if len(mark) == 0:
            print('No marks yet')
        else:
            for i , m in enumerate(mark, 1):
                print(f"{i}.{m}")
            
    elif choice == '3':
        if len(mark) == 0:
            print('no marks yet')
        else:
            avg = lambda x: sum(x) / len(x)
            print(f"Average {avg(mark):.1f}")

    elif choice == '4':
        if len(mark) ==0:
            print('NO MARKS YET')
        else:
            passed = list(filter(lambda x : x >= 40, mark))
            failed = list(filter(lambda x : x < 40, mark))
            print(f"passed : {len(passed)}")
            print(f"failed: {len(failed)}")

    elif choice == '5':
        print('goodbye')
        break 
    else:
        print('invalid')

    
```

```python
# test outside loop
mark = [91, 38, 78]
print("Test view:")
for i, m in enumerate(mark, 1):
    print(f"{i}. {m}")

avg = lambda x: sum(x) / len(x)
print(f"Average: {avg(mark):.1f}")

passed = list(filter(lambda x: x >= 40, mark))
failed = list(filter(lambda x: x < 40, mark))
print(f"Passed: {len(passed)}")
print(f"Failed: {len(failed)}")
```
