# Conditional statement 

You know —
✅ if / elif / else basic syntax
✅ conditions return True/False

Going deeper today —
⬜ nested conditions
⬜ ternary operator (one line if)
⬜ match statement (Python 3.10+)
⬜ short circuit evaluation
⬜ truthy/falsy in conditions

## syntax

```python
age = 23 

if age < 13:
    print('child')
elif age < 18:
    print('teenager')
elif age < 25:
    print('young adult')
elif age < 60:
    print('adult')
else:
    print('senior')
```

### Deeper 1  --> Nested condition (condition inside condition)

```python
age = 23 
salary = 35000 

if age >=21:
    if salary >= 25000:
        print('loan approved')
    else:
        print('salary is low')
else:
    print('age is low')
```

```python
# but nested gets messy , we use and / or instead 
if age>= 21 and salary >= 25000:
    print('loan approved')
elif age < 21:
    print('age to low')
else:
    print('salary to low')
```

### Deeper 2 ---> Ternary operator

```python
# normal

if marks >= 40:
    result = 'pass'
else:
    result = 'fail'

# ternary - 1 line 

result = "pass" if marks >= 40 else "fail"

#result = [value if True] if [condition] else [value if False]
```

## Problem 2 

```python
Take age and income as input
Check loan eligibility —
age must be 21-58 AND income >= 25000
Use short circuit — check age first
If age fails — don't check income at all
Print which condition failed
```

```python
age = int(input('enter your age'))
income = int(input('enter your income'))

if age >= 21 and age <= 58:
    if income >= 25000:
        print('loan eligibility')
    else:
        print(f'{income},income too low')
else:
    print(f'{age} too low')
```

## Problem 3 

```python
fruits = ["apple", "mango", "banana", "grapes"]

Take a fruit name as input
Check if it exists using 'in'
If exists → "fruit found ✅"
If not    → "fruit not found ❌ — adding it"
           → add to list
           → print updated list
```

```python
fruit = input('enter fruit name:')

fruits = ["apple", "mango", "banana", "grapes"]


if fruit in fruits:
    print(f'{fruit} found')
else:
    fruits.append(fruit)
    print('fruit not found')
   
print(fruits)


```

## problem - 5

```python
Build a simple chatbot model selector —
Ask user which model they want —
"gpt", "gemini", "llama", "mistral"

Use match statement —
gpt     → "Using OpenAI GPT ✅"
gemini  → "Using Google Gemini ✅"
llama   → "Using Local LLaMA ✅"
mistral → "Using Mistral ✅"
default → "Unknown model — using default"
```

```python
model = input('select any one model (gpt, gemini, llama, mistral, default) :')

match model:
    case "GPT":
        print('using open ai')
    case "gemini":
        print('using google gemini')
    case "llama":
        print('using local llama')
    case "mistral":
        print('using mistral')
    case "default":
        print('using default')
```
